## Scrape data of Laptops

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup


response = requests.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops")
print(response.status_code)

soup = BeautifulSoup(response.content, "html.parser")


product_cards = soup.find_all("div", class_= "col-md-4 col-xl-4 col-lg-4")

all_products = []

for card in product_cards:
    title_tag = card.find("a", class_= "title")
    title_name = title_tag.text.strip() if title_tag else "N/A"

    link = ("https://webscraper.io" + title_tag["href"] if title_tag else "N/A" )

    img_tag = card.find("img", class_ = "card-img-top")
    img =  "https://webscraper.io" + img_tag["src"] if img_tag else "N/A"

    price = card.find("h4", class_ = "price").text.strip()
    description = card.find("p", class_ = "description").text.strip()   
    reviews = card.find("p", class_ = "review-count").text.strip()
    all_products.append(
        {
            "Title": title_name,
            "Link": link,
            "Image": img,
            "Price": price,
            "Description": description,
            "Reviews": reviews
        }
    )


df = pd.DataFrame(all_products)
df.to_csv("laptops_data.csv", index=False)   
```
